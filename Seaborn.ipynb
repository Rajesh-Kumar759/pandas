```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Distribution Graph
# -------> 1. Displot(Distribution)
# -------> 2. Joinplot
# -------> 3. Pairplot
# -------> 4. Rugplot

#categorical Plots
# -------> 1. Barplot
# -------> 2. Countplot
# -------> 3. Boxplot
# -------> 4. Violinplot

#Matrix Plot
# -------> 1. Heatmap
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)   
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')   
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') 
```

```python
sns.pairplot(tips) #only numerical data,gives relation between all columns
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
tips
```

```python
sns.barplot(x='day',y='tip',data=tips)  #x-axis,y-axis,data,(mean is calculated)the black bar signifies Standard Deviation/Confidence interval
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='sex') #hue parameter gives additional parameter
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='smoker') 
```

```python
import numpy as np
sns.barplot(x='day',y='tip',data=tips,hue='smoker',estimator=np.median)   #here median is calculated
```

# Count plot 

```python
sns.countplot(tips['smoker'])  #used to give the count within a coulmn(here no of smokers and non-smokers)
```

```python
sns.countplot(tips['smoker'],data=tips,hue='sex')   #no of smoker in males and females
```

# Box plot

```python
#applied only on a numerical column
```

```python
sns.boxplot('total_bill',data=tips)
```

```python
sns.boxplot('total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x=tips['sex'],y=tips['tip'])  #for two columns(x-axis,y-axis)
```

```python
sns.violinplot(x=tips['sex'],y=tips['tip'])   #KDE
```

```python
sns.violinplot(x=tips['day'],y=tips['tip'])
```

# Heat map

```python
flight=sns.load_dataset('flights')
```

```python
flight.head()
```

```python
import pandas as pd
```

```python
p=flight.pivot_table(index='year',columns='month',values='passengers')
p
```

```python
sns.heatmap(p)
```

```python
sns.heatmap(p,annot=True)
```

```python
sns.heatmap(p,annot=True,linewidths=3)
```

```python
sns.heatmap(p,annot=True,linewidths=3,linecolor='black')
```

```python
sns.heatmap(p,annot=True,linewidths=3,cmap='summer')
```

```python
sns.heatmap(p,linewidths=3,linecolor='black',cmap='summer')
```

```python
sns.heatmap(p,linewidths=3,linecolor='black',cmap='autumn')
```

```python
sns.heatmap(p,linewidths=3,cmap='autumn')
```
